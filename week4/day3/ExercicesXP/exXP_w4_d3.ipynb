ðŸŒŸ **Exercise 3:** Understanding Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# complete your code here
mu, sig = 50, 10
data = np.random.normal(mu, sig, 1000)

# Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Plot the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

p = norm.pdf(x, mu, sig) # density proba
plt.plot(x, p, 'k', linewidth=2)
plt.title('Normal Distribution ($\mu=50$, $\sigma=10$)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

ðŸŒŸ **Exercise 4:** T-Test Application

```python
np.random.seed(0)

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# complete your code here
t_statistic, p_value = stats.ttest_ind(data1, data2)
# H0 : The means of the two samples are equal.
# H1 : The means of the two samples are different.
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

P-value < 0.05. This is a strong evidence to reject the null hypothesis.This suggest a significant difference between the two data set.

ðŸŒŸ **Exercise 5:** Linear Regression Analysis

```python
house_sizes = [50, 70, 80, 100, 120] # (in square meters)
house_prices = [150000, 200000, 210000, 250000, 280000] #(in currency units)

# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

linear_regression = [slope * x + intercept for x in house_sizes]

# Plotting the data and the linear regression line
plt.scatter(house_sizes, house_prices, color='b')
plt.plot(house_sizes, linear_regression, color='red')
plt.xlabel('House Size (sq. meters)')
plt.ylabel('House Price (currency units)')
plt.title('Linear Regression Analysis')
plt.grid(True, alpha=0.5)

plt.show()
```

- What is the slope and intercept of the regression line?

```python
# The slope and intercept of the regression line
print("Slope:", slope)
print("Intercept:", intercept)
```

- Predict the price of a house that is 90 square meters.

```python
price_90 = slope * 90 + intercept
print("Price of a house that is 90 square meters :", price_90)
```

- Interpret the meaning of the slope in the context of housing prices.


- Interpretation :
  - **slope** : This tells us how much the price increases for each additional square meter.
  - **intercept**: This represents the predicted price of a house with 0 square meters (which may not be meaningful in this case).
  - **r_value** : Indicates the strenght of the correlation.
  - **p_value** : Helps test if the relationship is statistically significant.
  - **std_err** : The standard error measures the accuracy of the slope estimate.

ðŸŒŸ **Exercise 6:** Understanding ANOVA

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
```

- Perform an ANOVA test on the given data. What are the F-value and P-value?

```python
# Performing one-way ANOVA
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value)
print("P-value:", p_value)
```

- **F-value** = 28.53 â†’ This tells us how much variation exists between the groups compared to within the groups.
- **P-value** = 2.75e-05 â†’ This represents the probability of getting these results if the fertilizers actually had no effect.

- Based on the P-value, do the fertilizers have significantly different effects on plant growth?

Since the P-value is much smaller than 0.05, we reject the null hypothesis.
This means that at least one of the fertilizers has significantly different effects on plant growth.

- Explain what would happen if the P-value were greater than 0.05.


If P-value > 0.05, we would fail to reject the null hypothesis. This would mean that there is no significant difference between the fertilizersâ€”in other words, the growth differences could just be due to random chance rather than the fertilizer type.

ðŸŒŸ **Exercise 7:** Probability Distributions (Optional)


Work with a binomial distribution: calculate probabilities for different numbers of successes.
Example: Calculating the probability of getting exactly 5 heads in 10 coin flips.

```python
from scipy.stats import binom

# Define parameters
n = 10  # Number of trials (flips)
p = 0.5  # Probability of success (heads)
k = 5  # Number of desired heads

# Compute the probability of exactly 5 heads
prob_5_heads = binom.pmf(k, n, p)
print(f"Probability of getting exactly 5 heads in 10 flips: {prob_5_heads:.4f}")

# Plot the binomial distribution
x_binom = np.arange(0, n+1)  # Possible number of heads (0 to 10)
pmf_binom = binom.pmf(x_binom, n, p)  # Binomial probability mass function

plt.figure(figsize=(8, 5))
plt.bar(x_binom, pmf_binom, color='royalblue', alpha=0.7, label="PMF")
plt.axvline(k, color='red', linestyle="--", label="k = 5")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

ðŸŒŸ **Exercise 8:** Correlation Coefficients (Optional)

Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.

```python
import pandas as pd

data = pd.DataFrame({'age': [23, 25, 30, 35, 40],
                     'income': [35000, 40000, 50000, 60000, 70000]})
```

**Pearson Correlation Coefficient:**

The Pearson Correlation Coefficient is a statistical measure that calculates the strength and direction of the linear relationship between two continuous variables. Its value ranges from -1 to 1, where:
- 1 indicates a perfect positive linear relationship,
- -1 indicates a perfect negative linear relationship, and
- 0 indicates no linear relationship.

```python
pearson_corr, _ = stats.pearsonr(data['age'], data['income'])
print(f"Pearson Correlation Coefficient: {pearson_corr:.2f}")
```

**Spearman Correlation Coefficient:**

 The Spearman correlation evaluates how well the relationship between two variables can be described using a monotonic function (not necessarily linear), where the variables tend to move in the same direction but not necessarily at a constant rate.
 Instead of using raw values, Spearman uses the ranks of the data points.

```python
spearman_corr, _ = stats.spearmanr(data['age'], data['income'])
print(f"Spearman Correlation Coefficient: {spearman_corr:.2f}")
```
