```python
import numpy as np

# Example survey data matrix (5 respondents, 4 questions)
survey_data = np.array([[5, 1, 4, 4],
                        [4, 2, 4, 3],
                        [5, 1, 1, 5],
                        [3, 2, 1, 2],
                        [2, 1, 2, 5]])

# Calculating various statistics
average_responses = np.mean(survey_data, axis=0)
total_responses = np.sum(survey_data, axis=0)
max_responses = np.max(survey_data, axis=0)
min_responses = np.min(survey_data, axis=0)

print("Average Responses:", average_responses)
print("Total Responses:", total_responses)
print("Maximum Responses:", max_responses)
print("Minimum Responses:", min_responses)
```

```python
import numpy as np
import pandas as pd
import time

# Creating a large DataFrame
data = pd.DataFrame(np.random.rand(100, 10), columns=['A', 'N', 'C', 'D','E','N', 'B', 'C', 'D','E'])
data.head()
```

```python
# Timing Pandas operations
start_time = time.time()
result_pandas = data[(data['A'] > 0.5) & (data['B'] < 0.5)].mean()
end_time = time.time()
print("Pandas Conditional Mean:\n", result_pandas)
print("Pandas Operation Time: {:.5f} seconds".format(end_time - start_time))
```

```python
# Timing equivalent NumPy operations
numpy_data = data.to_numpy()
start_time = time.time()
result_numpy = np.mean(numpy_data[(numpy_data[:, 0] > 0.5) & (numpy_data[:, 1] < 0.5)], axis=0)
end_time = time.time()
print("NumPy Conditional Mean:\n", result_numpy)
print("NumPy Operation Time: {:.5f} seconds".format(end_time - start_time))
```

```python
import matplotlib.pyplot as plt

# Plotting with NumPy arrays
plt.scatter(numpy_data[:, 0], numpy_data[:, 1], alpha=0.5)
plt.title("Custom Scatter Plot")
plt.xlabel("A")
plt.ylabel("B")
plt.show()
```

# **Exercises on Time Series Data in NumPy- solution**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Data Creation

# Generate date range
date_range = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')

# Simulate temperature and humidity data
temperature = np.random.uniform(10, 35, len(date_range))
humidity = np.random.uniform(30, 90, len(date_range))

# Create DataFrame
df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})
print(df.head())

```

```python
# 2. Data Analysis

# Calculate statistics
mean_temp, median_temp, std_temp = df['Temperature'].mean(), df['Temperature'].median(), df['Temperature'].std()
mean_humidity, median_humidity, std_humidity = df['Humidity'].mean(), df['Humidity'].median(), df['Humidity'].std()

print("Temperature Stats:", f"Mean: {mean_temp:.2f}, Median: {median_temp:.2f}, Std Dev: {std_temp:.2f}")
print("Humidity Stats:", f"Mean: {mean_humidity:.2f}, Median: {median_humidity:.2f}, Std Dev: {std_humidity:.2f}")

#  highest temperature and humidity dates
max_temp_date = df.loc[df['Temperature'].idxmax(), 'Date']
max_humidity_date = df.loc[df['Humidity'].idxmax(), 'Date']

print(f"Date with highest temperature: {max_temp_date}")
print(f"Date with highest humidity: {max_humidity_date}")

# Plot temperature and humidity trends
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Temperature'], label='Temperature (C)', color='r')
plt.plot(df['Date'], df['Humidity'], label='Humidity (%)', color='b')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Daily Temperature and Humidity Trends')
plt.legend()
plt.show()
```

**# 3.Rolling analesis With Window Function**

```python


#  7-day moving average
df['Temp_MA_7'] = df['Temperature'].rolling(window=7).mean()
df['Humidity_MA_7'] = df['Humidity'].rolling(window=7).mean()

# Plot original vs. moving average
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Temperature'], label='Original Temperature', color='r', alpha=0.5)
plt.plot(df['Date'], df['Temp_MA_7'], label='7-Day Moving Avg (Temp)', color='r')
plt.plot(df['Date'], df['Humidity'], label='Original Humidity', color='b', alpha=0.5)
plt.plot(df['Date'], df['Humidity_MA_7'], label='7-Day Moving Avg (Humidity)', color='b')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Temperature and Humidity with 7-Day Moving Average')
plt.legend()
plt.show()
```
