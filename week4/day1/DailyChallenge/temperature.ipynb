```python
import numpy as np
import random
import pandas as pd
```

1. **Data Preparation:**




```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
cities = ['city 1', 'city 2', 'city 3', 'city 4', 'city 5', 'city 6', 'city 7', 'city 8', 'city 9', 'city 10']
```

```python
# Set a random seed for reproducibility
np.random.seed(42)

data = np.round(np.random.uniform(-5, 35, size=(10,12)), 2)
```

```python
print(data)
```

```python
dataset = pd.DataFrame(data, index=cities, columns=months)
print(dataset)
```

2. **Data Analysis:**

```python
# avg per month
avg_temp_month = dataset.mean(axis=0)
print(avg_temp_month)
```

```python
# avg per city
avg_temp_city = dataset.mean(axis=1)
print(avg_temp_city)
```

```python
# annual avg
avg_year = avg_temp_city.mean()
print(avg_year)
```

```python
highest_temp_city = avg_temp_city.idxmax()
print(highest_temp_city)
```

```python
lowest_temp_city = avg_temp_city.idxmin()
print(lowest_temp_city)
```

3. **Data Visualization:**

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))

for city in cities:
    plt.plot(dataset.columns, dataset.loc[city], label=city, linestyle=random.choice(['-', '--', '-.', ':']))

plt.xlabel('Month', fontsize=12)
plt.ylabel('Temperature (Â°C)', fontsize=12)
plt.title('Monthly Temperature Trends for Each City', fontsize=14)
plt.legend(title="City", loc='lower left')

plt.show()
```
