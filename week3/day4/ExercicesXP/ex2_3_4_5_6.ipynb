ðŸŒŸ **Exercise 3:** Visualizing Monthly Sales with a Bar Chart

```python
sales = [5000, 5500, 6200, 7000, 7500, 8000, 7500, 8500, 7500, 8500, 6500, 7000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.bar(months, sales, edgecolor='purple', color='pink', hatch='/', alpha=0.6)

plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales Amount ($)', fontsize=12)
plt.title('Monthly Sales', fontsize=14, fontweight='bold')

plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.6)  # Add horizontal grid lines

plt.show()
```

ðŸŒŸ **Exercise 4:** Visualizing the Distribution of CGPA

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic CGPA data for 200 students (values between 2.0 and 4.0)
df = pd.DataFrame({'CGPA': np.random.uniform(2.0, 4.0, 200)})

sns.histplot(df['CGPA'], bins=20, color='skyblue', edgecolor='black')

plt.xlabel('CGPA', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.title('Distribution of CGPA Among Students', fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)  # Add a light grid for better visibility

plt.show()
```

ðŸŒŸ **Exercise 5:** Comparing Anxiety Levels Across Different Genders

```python
data = {
    'Do you have Anxiety?': ['Yes', 'No', 'No', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes'],
    'Choose your gender': ['Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Female', 'Male', 'Female', 'Male']
}

df = pd.DataFrame(data)

# Calculate the proportion of students with anxiety per gender
anxiety_proportions = df.groupby('Choose your gender')['Do you have Anxiety?'].apply(lambda x: (x == 'Yes').mean()).reset_index()

# Rename the column for clarity
anxiety_proportions.columns = ['Gender', 'Anxiety Proportion']

sns.barplot(x='Gender', y='Anxiety Proportion', data=anxiety_proportions, hue='Gender', palette='pastel')

plt.xlabel('Gender', fontsize=12)
plt.ylabel('Proportion of Students with Anxiety', fontsize=12)
plt.title('Anxiety Levels Across Different Genders', fontsize=14, fontweight='bold')
plt.ylim(0, 1)  # Set the y-axis limits to [0, 1]

plt.show()
```

ðŸŒŸ **Exercise 6:** Exploring the Relationship Between Age and Panic Attacks

```python
data = {
    'Age' : [18, 65, 20, 43, 90, 12, 84, 95, 56, 87, 23, 69, 98, 14, 77],
    'Do you have panic attack?': ['No', 'No', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes']
}
df = pd.DataFrame(data)

df['Panic Attack'] = df['Do you have panic attack?'].map({'Yes': 1, 'No': 0})

sns.scatterplot(x='Age', y='Panic Attack', data=df, color='purple', edgecolor='black', s=80, alpha=0.7)

plt.xlabel('Age', fontsize=12)
plt.ylabel('Panic Attack (Yes=1, No=0)', fontsize=12)
plt.title('Relationship Between Age and Panic Attacks', fontsize=14, fontweight='bold')

plt.show()
```
