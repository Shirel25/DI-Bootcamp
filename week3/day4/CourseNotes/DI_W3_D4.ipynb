```python
import matplotlib.pyplot as plt

activities = ['Work', 'Sleep', 'Leisure', 'Exercise', 'Meals', 'Others']
time_spent = [8, 7, 4, 2, 2, 1]  # Total = 24 hours

# Define custom colors
colors=['red', 'green', 'blue', 'purple','gray','black']

explode = (0, 0, 0, 0.4, 0, 0)


plt.pie(time_spent, labels=activities, autopct='%1.1f%%', colors=colors,
        explode=explode, shadow=True, startangle=140, wedgeprops={'edgecolor': 'black'})

# Add title
plt.title('Daily Time Distribution', fontsize=14, fontweight='bold')

# Show the plot
plt.show()
```

# **Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
data = sns.load_dataset('tips')
sns.histplot(data['tip'], bins=200, color='skyblue')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], shade=True, color='purple')
plt.title('KDE of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='scatter', color='green')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
sns.barplot(x='class', y='survived', data=titanic)
plt.title('Survival Rate by Class')
plt.show()
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='dark')
plt.title('Age Distribution by Class')
plt.show()
```

```python
sns.violinplot(x='class', y='age', data=titanic, palette='muted')
plt.title('Age Distribution by Class')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=data, aspect=1.5, height=5)
plt.title('Scatter Plot with Regression Line')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='reg', color='orange')
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species', palette='Dark2')
plt.show()
```

```python
correlation_matrix = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr() # Select only numerical columns for correlation

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
data = sns.load_dataset('tips')
g = sns.FacetGrid(data, col='time', row='smoker', margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip', color='red')
plt.show()
```

```python
sns.set_context('talk')
sns.set_style('whitegrid')
sns.histplot(data['total_bill'], color='darkblue', bins=15)
plt.title('Advanced Aesthetic Customization')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(0, 10, 100)  # Generate 100 points between 0 and 10
y = x**2  # Compute y = x^2

# Create the plot
plt.plot(x, y, label=r'$y = x^2$', color='b', linestyle='-', linewidth=2)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $y = x^2$')

# Show grid and legend
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.figure
import matplotlib.backends.backend_agg as agg

fig = matplotlib.figure.Figure()
agg.FigureCanvasAgg(fig)  # Required to render the figure in some backends
ax = fig.add_subplot(111)  # Manually adding a subplot
ax.plot([1, 2, 3], [1, 4, 9])
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Using Matplotlib Without Pyplot")

fig.savefig("plot.png")  # Need to explicitly save the figure
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
plt.figure(figsize=(8, 5))
sns.histplot(tips["total_bill"], bins=20, color="skyblue", edgecolor="black")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Histogram of Customer Spending")
plt.show()

```

```python

iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.show()

```

```python
sns.heatmap(iris.drop(columns="species").corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation  of Iris Dataset")
plt.show()

```

```python
titanic = sns.load_dataset("titanic")
g = sns.FacetGrid(titanic, col="sex", row="class", hue="survived", height=3, aspect=1.5)
g.map(sns.histplot, "age", bins=20, alpha=0.7)
g.add_legend()
plt.show()
```
