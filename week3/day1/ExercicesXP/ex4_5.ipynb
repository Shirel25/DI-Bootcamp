```python
import pandas as pd

iris = pd.read_csv('Iris Dataset/Iris.csv')
iris.head()
```

- **Quantitative** :
                      *   Id (int64)
                      *   SepalLengthCm (float64)
                      *   SepalWidthCm (float64)
                      *   PetalLengthCm (float64)
                      *   PetalWidthCm (float64)
Quantitative data can be used for statistical modeling and predictions.

- **Qualitative** :
                      *   Species (object)
Qualitative data helps in classification and pattern recognition.


```python
quantitative_colums = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]


print("Mean :\n", iris[quantitative_colums].mean())
print("\nMedian :\n", iris[quantitative_colums].median())
print("\nMode :\n", iris[quantitative_colums].mode())
# print("\nMode :\n", iris[quantitative_colums].mode().iloc[0]) take the first mode value
```

```python
import matplotlib.pyplot as plt

# Scatter plot of Sepal dimensions (Length vs Width)
plt.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'], label='Sepal', color='blue')

# Scatter plot of Petal dimensions (Length vs Width)
plt.scatter(iris['PetalLengthCm'], iris['PetalWidthCm'], label='Petal', color='red')

# Adding labels for axes
plt.xlabel('Length (cm)')
plt.ylabel('Width (cm)')

# Adding a title to describe the plot
plt.title("Scatter plot of Sepal and Petal Dimensions")

# Adding a legend to distinguish between Sepal and Petal data points
plt.legend()

# Displaying the plot
plt.show()
```

Findings:
- Sepal and petal measurements form distinct clusters.
- Petal dimensions (red points) are more compact and concentrated at lower values.
- Sepal dimensions (blue points) are more spread out, indicating greater variation.
- There is a noticeable positive correlation between length and width in both sepals and petals.
- This visualization helps to distinguish between different types of flowers based on their sepal and petal sizes.
