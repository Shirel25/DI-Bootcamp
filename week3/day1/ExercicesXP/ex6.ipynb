```python
import pandas as pd
data_americans = pd.read_csv('how-much-sleep-do-americans-really-get.csv')
data_americans.head()
```

### ***Analyzes***



```python
import matplotlib.pyplot as plt

sleep_trend = data_americans.groupby("Year")["Avg hrs per day sleeping"].mean()

plt.plot(sleep_trend.index, sleep_trend.values, marker='o', linestyle='-', color='blue', label="Average Sleep Duration")

plt.xlabel("Year")
plt.ylabel("Avg hrs per day sleeping")
plt.title("Trend of Sleep Duration Over the Year")
plt.show()
```

1. **Sleep Trend Over Time**
   - The `Year` column allows us to see if Americans are sleeping more or less over the years.
   - The sleep duration has increased over time, from around 8.70 hours in 2004 to nearly 8.90 in 2016.
   - The general trend is positive, indicating a possible improvement in sleep habits.

   **Potential reasons,for this increase :**
     - Greater awarness of sleep importance
     - Changes in lifestyle (more remote work, different work schedules)


```python
import seaborn as sns

sns.boxplot(x="Sex", y="Avg hrs per day sleeping", data=data_americans)
plt.title("Comparison of Sleep Duration by Gender")
plt.show()
```

2. **Sleep Duration by Gender**
   - This boxplot compares the **average hours of sleep per day** among **Men, Women, and Both genders combined**.
   - The **median sleep duration** appears to be slightly higher for **Women** compared to **Men**, but the difference is minimal.
   - The **spread of data** is similar across all three groups, meaning **sleep patterns are distributed in a similar way** for both genders.
   - The majority of people sleep between **8 and 9 hours per night**, as indicated by the central boxes.
