```python
import pandas as pd

ds_job_salary = pd.read_csv('datascience_salaries.csv')
ds_job_salary.head()
```

```python
print(ds_job_salary['salary'].describe())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
ds_job_salary['salary_normalized'] = scaler.fit_transform(ds_job_salary[['salary']])
# []   -> Series
# [[]] -> dataFrame

# ds_job_salary.head()
print(ds_job_salary[['salary', 'salary_normalized']].head())
```

```python
from sklearn.decomposition import PCA

#This line creates an instance of the PCA class with the parameter n_components set to 2.
# It specifies that we want to reduce the dimensionality of the data to 2 principal components.
#In other words, the data will be projected into a 2D space.
pca = PCA(n_components=1)
reduced_ds_job_salary = pca.fit_transform(ds_job_salary[['salary_normalized']])

ds_job_salary["PCA_Salary"] = reduced_ds_job_salary
print(ds_job_salary.head())
```

```python
grp_experience_level = ds_job_salary.groupby('experience_level')['salary'].agg(['mean', 'median'])
print(grp_experience_level)
```
